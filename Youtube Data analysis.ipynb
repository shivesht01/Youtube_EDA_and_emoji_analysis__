# Sentiment Polarity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
comments=pd.read_csv('F:\Python Data Analysis\Youtube Text Data Analysis/GBcomments.csv', error_bad_lines=False ,low_memory=False)
```

```python
comments.head()
```

```python
!pip install Textblob
```

```python
from textblob import TextBlob
```

```python
TextBlob("It's more accurate to call it the M+ (1000) be..").sentiment.polarity
```

```python
comments.isna().sum()
```

```python
comments.dropna(inplace=True)
```

```python
sentiment_polarity=[]
for i in comments['comment_text']:
    sentiment_polarity.append(TextBlob(i).sentiment.polarity)
    
```

```python
comments['Sentiment_Polarity']=sentiment_polarity
```

```python
comments.head()
```

# EDA of positive and negetive Comments

```python
positive_sentiments=comments[comments['Sentiment_Polarity']==1]
```

```python
positive_sentiments.head()
```

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
combined_comments=''.join(positive_sentiments['comment_text'])
```

```python
wordcloud=WordCloud(height=500, width=1000, stopwords=stopwords).generate(combined_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

```python
comments_negetive=comments[comments['Sentiment_Polarity']==-1]
```

```python
comments_negetive.head()
```

```python
combined_comments=''.join(comments_negetive['comment_text'])
```

```python
wordcloud=WordCloud(width=1000, height=500, stopwords=stopwords).generate(combined_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

# EDA of used Tags

```python
videos=pd.read_csv('F:\Python Data Analysis\Youtube Text Data Analysis/USvideos.csv', error_bad_lines=False)
```

```python
videos.head()
```

```python
tags_combined=''.join(videos['tags'])
```

```python
import re
```

```python
tags=re.sub('[^a-zA-Z]',' ',tags_combined)
```

```python
tags
```

```python
tags=re.sub(' +',' ',tags)
```

```python
wordcloud=WordCloud(width=1000, height=500, stopwords=stopwords).generate(tags)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

# Regression Plot

```python
sbn.regplot(data=videos, x='views', y='likes')
plt.title("Views vs Like Regression Graph")
```

```python
sbn.regplot(data=videos, x='views', y='dislikes')
plt.title("Views vs Dislikes regression Plot")
```
